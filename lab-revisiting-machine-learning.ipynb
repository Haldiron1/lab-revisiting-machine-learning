```python
import pandas as pd
```

```python
df= pd.read_csv("learningSet.csv")
```

```python
df
```

```python
null_counts = df.isnull().sum()
```

```python
print(null_counts[null_counts > 0])
```

```python
exclude_vars = ['OSOURCE', 'ZIP CODE']

# Create an empty list to store columns to be dropped
drop_list = []

# Append columns to drop_list
for var in exclude_vars:
    if var in df.columns:
        drop_list.append(var)

# Drop the columns from the DataFrame
df = df.drop(columns=drop_list)

# Now df contains the modified DataFrame with the specified columns dropped
```

```python
df
```

```python
# Calculate the percentage of missing values for each column
missing_percent = (df.isna().sum() / len(df)) * 100

# Identify columns with over 85% missing values
columns_over_85_percent_missing = missing_percent[missing_percent > 85].index.tolist()

# Print the identified columns
print(columns_over_85_percent_missing)
```

```python
df = df.drop(columns=columns_over_85_percent_missing)
```

```python
df
```

```python
gender_mapping = {'M': 'M','F': 'F'}

# Apply the mapping, and set 'other' for remaining categories
df['GENDER'] = df['GENDER'].map(gender_mapping).fillna('other')


```
